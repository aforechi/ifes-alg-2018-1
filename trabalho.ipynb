## Descrição do trabalho: 
Implementar um programa em C/C++ que leia do [arquivo](./trabalho/entrada.ply) um conjunto de pontos 3D [1] e suas cores correspondentes no formato RGB [2] **e armazene em duas matrizes separadas, sendo a primeira de float utilizando o TAD [tMatrizFloat](./trabalho/matrizfloat.h) e a segunda utilizando o TAD [tMatrizInt](./trabalho/matrizint.h)**. Após a leitura completa do arquivo, faça o processamento desses pontos e cores da seguinte forma:
- translação (usando multiplicação de matriz) de todos os pontos 3D em uma ou mais coordenadas simultâneamente;
- mudança de escala (usando multiplicação de matriz) de todos os pontos 3D por um fator positivo real para cada coordenada x, y e z;
- conversão das cores dos pontos para escala de cinza segundo a fórmula [5];
- rotação (usando multiplicação de matriz) de todos os pontos 3D em um ou mais eixos simultâneamente [4];
- projeção (usando multiplicação de matriz) de todos os pontos 3D em um plano xy, xz, yz;

Implemente cada operação de processamento de cores e pontos em funções específicas usando matrizes<s>, alocação dinâmica e apontadores</s>. E salve o resultado do processamento em arquivo texto conforme o padrão [3].

As opções do programa são as seguintes:
- Digite a letra E para mudar a escala dos pontos 3D seguido de um fator de escala maior que zero para cada coordenada x y z;
- Digite a letra T para translação dos pontos 3D seguido das distâncias de translação x y z desejadas;
- Digite a letra R para rotacionar os pontos 3D seguido dos ângulos (em graus) de rotação em x, y e z (importante: permitir rotação simultânea dos três eixos)
- Digite a letra C para converter a cor dos pontos 3D para tons de cinza.
- Digite a letra P para projeção dos pontos 3D seguido do eixo descartado para a projeção no plano composto pelos outros dois eixos. Ex: digite z para projetar os pontos no plano xy.
- Digite a letra S para salvar as transformações em arquivo (.ply) e finalizar a execução do programa com sucesso.

### Convenções:
1.	Um ponto tridimensional (3D) é representado por três coordenadas x, y, z do plano cartesiano, convencionadas nesta ordem. Onde cada coordenada deverá ser representada por um número real (ponto flutuante em C).
2.	Uma cor pode ser representada pela composição de três cores principais vermelho (red, em inglês), verde (green, em inglês) e azul (blue, em inglês) comumente conhecido como padrão RGB de cores. Onde cada uma das três cores principais do padrão RGB pode assumir valores inteiros entre 0 e 255, respectivamente, dos tons mais escuros aos mais claros. Por exemplo: a cor branca é representada com os componentes RGB todos iguais a 255 e a cor preta com todos os valores de R,G,B iguais a 0. Outros valores de RGB iguais representam diferentes tons de cinza, desde o mais escuro ao mais claro.
3.	O formato de entrada e saída de dados do programa para pontos e cores deve respeitar a seguinte ordem para leitura e exibição de cada ponto e cor por linha: x y z r g b
4.	Conversão de cor RGB para tons de cinza calculando a média dos canais RGB e substituindo os valores de RGB pela média calculada para cada cor. c = (r + g + b)/3
5.	O sistema de coordenadas para 3D utilizado será o da Regra da Mão Direita, com o eixo Z perpendicular ao papel e saindo em direção ao observador.
    - O sentido positivo de uma rotação é dado quando observando-se sobre um eixo positivo em direção à origem, uma rotação de 90° irá levar um eixo positivo em outro positivo.
